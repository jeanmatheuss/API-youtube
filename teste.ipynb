```python
import pandas as pd
from datetime import datetime

# Exemplo de DataFrame
data = {
    "Título": ["Vídeo 1", "Vídeo 2", "Vídeo 3"],
    "Visualizações": [1000, 2000, 3000],
    "Data de Publicação": ["2023-09-20T10:00:00Z", "2023-09-19T15:30:00Z", "2023-09-18T08:45:00Z"]
}

df = pd.DataFrame(data)

# Converte a coluna "Data de Publicação" para um objeto de data
df["Data de Publicação"] = pd.to_datetime(df["Data de Publicação"])

# Calcula a diferença em horas entre a data atual e a data de publicação
current_time = datetime.now()
df["Horas desde Publicação"] = (current_time - df["Data de Publicação"]).dt.total_seconds() / 3600

# Exibe o DataFrame com a nova coluna
print(df)
```

```python
# função para fazer update diário dos novos vídeos
def update_df(df, df_up):
    df = pd.concat([df,df_up], ignore_index=True)
    return df
```

```python
import pandas as pd

# DataFrame original
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)

# DataFrame de atualização
data_up = {'A': [7, 8, 9], 'B': [10, 11, 12]}
df_up = pd.DataFrame(data_up)

# Chamando a função para atualizar df com df_up
df1 = update_df(df, df_up)

# Exibindo o DataFrame atualizado
df1
```
